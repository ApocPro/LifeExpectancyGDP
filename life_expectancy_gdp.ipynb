```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

life_expectancy = pd.read_csv("all_data.csv")

life_expectancy.tail(20)
```

```python
life_expectancy['Country'].value_counts()
```

# Project Scope

Looking at our database this is a limited amount of data.  What we want to do is come up with some useful visualisations that demonstrate any information that has been collected here.

Country to Life expectancy = Category to quatitative
Year to Life expectancy = Ordinal to quantitative
GDP to Life expectancy = Quantitative to Quantitative

```python
sns.scatterplot(x = 'GDP', y = 'Life expectancy at birth (years)', hue = 'Country', data = life_expectancy)
```

```python
Zimbabwe = life_expectancy.loc[life_expectancy['Country'] == 'Zimbabwe']
sns.lmplot(x = 'GDP', y = 'Life expectancy at birth (years)', data = Zimbabwe, line_kws={'color': 'black'}, lowess=True)
```

```python
the_rest = life_expectancy.loc[life_expectancy['Country'] != 'Zimbabwe']
sns.scatterplot(x = 'GDP', y = 'Life expectancy at birth (years)', hue = 'Country', data = the_rest)
```
